## Linear Regression Algorithm 01: (Y axis is taken where x=0)
        1. Define a random line slope = 1, pivot = 0
        2. Select number of iterations = 10000, rate of change = 0.01
        3. Train the line with sample points
            Case1: Point is above the line and right side of y-axis
                slope += slope*rate_of_change
                pivot += pivot*rate_of_change
            Case2: Point is above the line and left side of y-axis
                slope += slope*rate_of_change
                pivot -= pivot*rate_of_change 
            Case3: Point is below the line and right side of y-axis
                slope += slope*rate_of_change
                pivot -= pivot*rate_of_change
            Case4: Point is below the line and left side of y-axis
                slope -= slope*rate_of_change
                pivot += pivot*rate_of_change 
        4. Plot the trained line
                    

```python
'''
Liner regression:
This is a program to simulate Linear regression algorithm
'''
import random
import matplotlib.pyplot as plot


class Line:
    '''
    Class: Line
    Member functions: add_points, get_output, plot_graph and train_Line
    '''

    def __init__(self, rate=0.01, iterations=1000, slope=0, pivot=0):
        self.slope = slope
        self.pivot = pivot
        self.points = []
        self.iterations = iterations
        self.rate = rate
        self.noofpoints = 0
        self.__trained = False

    def add_points(self, list1):
        self.points = list1
        self.noofpoints = len(self.points)

    def get_output(self, x):
        return self.pivot + self.slope * x

    def plot_graph(self):
        x = list(range(-10, 10))
        y = list(map(self.get_output, x))

        points_x = [x[0] for x in self.points]
        points_y = [x[1] for x in self.points]

        fig = plot.figure()
        ax = fig.add_subplot(1, 1, 1)
        ax.set_xlim(-10,10)
        ax.set_ylim(-10,50)


        # Move left y-axis and bottom x-axis to centre, passing through (0,0)
        ax.spines['left'].set_position('center')
        ax.spines['bottom'].set_position('center')

        # Eliminate upper and right axes
        ax.spines['right'].set_color('none')
        ax.spines['top'].set_color('none')

        # Show ticks in the left and lower axes only
        ax.xaxis.set_ticks_position('bottom')
        ax.yaxis.set_ticks_position('left')

        plot.plot(x, y, points_x, points_y, 'r^')

        # Show ticks in the left and lower axes only
        ax.xaxis.set_ticks_position('bottom')
        ax.yaxis.set_ticks_position('left')

        if self.__trained == False:
            plot.title("Linear regression: Before training")
        else:
            plot.title("Linear regression: After training")

        plot.show()

    def train_Line(self):
        #         print("Going to train Line")
        for i in range(self.iterations):
            for j in range(len(self.points)):
                x = self.points[j][0]
                act = self.points[j][1]

                y = self.pivot + self.slope * x
                if act > y:
                    self.pivot += self.rate
                    if x >= 0:
                        self.slope += self.rate
                    else:
                        self.slope -= self.rate
                elif act < y:
                    self.pivot -= self.rate
                    if x >= 0:
                        self.slope -= self.rate
                    else:
                        self.slope += self.rate
        self.__trained = True
        print("Training is over....:)")


'''
Creating object of class <Line> : Linear regression algorith.
The following are the argument list and its default values.
rate=0.1 ---> Rate of change of slope for each oteration
iterations=100  ---> Number of iterations to train the list.
slope=0 ---> Initial value of slope
pivot=1 ---> Initial value of pivot
'''
Line1 = Line()  # Creating object
Line1.add_points([(-7,-3), (-6,0),(-5,2),(-4,4),(-2,20), (-1,40),(0,45),(1,50)])#Add two points to train
Line1.plot_graph()  # Draw the line: before training
Line1.train_Line()  # Train the line with given points
Line1.plot_graph()  # Draw the line: after training

Line1 = Line()  # Creating object
Line1.add_points([(-7,-3), (-6,0),(-5,2),(-4,4),(-2,20), (-1,40)])#Add two points to train
Line1.plot_graph()  # Draw the line: before training
Line1.train_Line()  # Train the line with given points
Line1.plot_graph()  # Draw the line: after training
```

# Linear Regression Algorithm 02:
    1. Define a random line slope = 1, pivot = 0
    2. Select number of iterations = 10000, rate of change = 0.01
    3. Train the line with sample points
                self.slope += (act - y)*(x)*self.rate
                self.pivot += (act - y)*self.rate
        (act - y) -----> Distance between line and y.
                x -----> Distance between y-axis and x.

```python
'''
Liner regression:
This is a program to simulate Linear regression algorithm
'''
import random
import matplotlib.pyplot as plot
from numpy import median

class Line:
    '''
    Class: Line
    Member functions: add_points, get_output, plot_graph and train_Line
    '''

    def __init__(self, rate=0.01, iterations=1000, slope=0, pivot=0):
        self.slope = slope
        self.pivot = pivot
        self.points = []
        self.iterations = iterations
        self.rate = rate
        self.noofpoints = 0
        self.__trained = False

    def add_points(self, list1):
        self.points = list1
        self.noofpoints = len(self.points)
        self.x_values = [x[0] for x in self.points]
        self.x_axis = median(self.x_values)

    def get_output(self, x):
        return self.pivot + self.slope * x

    def plot_graph(self):
        x = list(range(-10, 10))
        y = list(map(self.get_output, x))
        #         plot.ylim((-20, 10))

        points_x = [x[0] for x in self.points]
        points_y = [x[1] for x in self.points]

        fig = plot.figure()
        ax = fig.add_subplot(1, 1, 1)
        ax.set_xlim(-10,10)
        ax.set_ylim(-10,50)

        # Move left y-axis and bottom x-axis to centre, passing through (0,0)
        ax.spines['left'].set_position('center')
        ax.spines['bottom'].set_position('center')

        # Eliminate upper and right axes
        ax.spines['right'].set_color('none')
        ax.spines['top'].set_color('none')

        # Show ticks in the left and lower axes only
        ax.xaxis.set_ticks_position('bottom')
        ax.yaxis.set_ticks_position('left')

        plot.plot(x, y, points_x, points_y, 'r^')

        # Show ticks in the left and lower axes only
        ax.xaxis.set_ticks_position('bottom')
        ax.yaxis.set_ticks_position('left')

        if self.__trained == False:
            plot.title("Linear regression: Before training")
        else:
            plot.title("Linear regression: After training")

        plot.show()

    def train_Line(self):
        for i in range(self.iterations):
            for j in range(len(self.points)):
                x = self.points[j][0]
                act = self.points[j][1]
                y = self.pivot + self.slope * x
                
                self.slope += (act - y)*(x)*self.rate
                self.pivot += (act - y)*self.rate
        self.__trained = True
        print("Training is over....:)")


'''
Creating object of class <Line> : Linear regression algorith.
The following are the argument list and its default values.
rate=0.1 ---> Rate of change of slope for each oteration
iterations=100  ---> Number of iterations to train the list.
slope=0 ---> Initial value of slope
pivot=1 ---> Initial value of pivot
'''
Line1 = Line()  # Creating object
Line1.add_points([(-7,-3), (-6,0),(-5,2),(-4,4),(-2,20), (-1,40),(0,45),(1,50)])#Add two points to train
Line1.plot_graph()  # Draw the line: before training
Line1.train_Line()  # Train the line with given points
Line1.plot_graph()  # Draw the line: after training



Line1 = Line()  # Creating object
Line1.add_points([(-7,-3), (-6,0),(-5,2),(-4,4),(-2,20)])#Add two points to train
Line1.plot_graph()  # Draw the line: before training
Line1.train_Line()  # Train the line with given points
Line1.plot_graph()  # Draw the line: after training
```
